```python
def calculate_simple_interest():
    print("\n--- Simple Interest Calculation ---")
    P = float(input("Enter Principal amount (P): "))
    R = float(input("Enter Rate in percentage (R): "))
    T = float(input("Enter Time in years (T): "))
    
    # Formula: A = P(1 + (R/100)T)
    A = P * (1 + (R / 100) * T)
    print(f"Total Amount (A): {A:.2f}")

def calculate_compound_interest():
    print("\n--- Compound Interest Calculation ---")
    P = float(input("Enter Principal amount (P): "))
    R = float(input("Enter Annual Rate as a decimal (R, e.g., 0.05 for 5%): "))
    n = float(input("Enter number of times interest is compounded per year (n): "))
    t = float(input("Enter Time in years (t): "))
    
    # Formula: A = P(1 + R/n)^(nt)
    A = P * (1 + R / n) ** (n * t)
    print(f"Total Amount (A): {A:.2f}")

def calculate_annuity_plan():
    print("\n--- Annuity Plan Calculation ---")
    PMT = float(input("Enter Payment Amount per period (PMT): "))
    R = float(input("Enter Annual Rate as a decimal (R, e.g., 0.05 for 5%): "))
    n = float(input("Enter number of times interest is compounded per year (n): "))
    t = float(input("Enter Time in years (t): "))
    
    # Formula: A = PMT * [((1 + R/n)^(nt) - 1) / (R/n)]
    numerator = (1 + R / n) ** (n * t) - 1
    denominator = R / n
    A = PMT * (numerator / denominator)
    print(f"Total Annuity Amount (A): {A:.2f}")

def main():
    while True:
        print("\n=== Financial Calculator Menu ===")
        print("1. Calculate Simple Interest")
        print("2. Calculate Compound Interest")
        print("3. Calculate Annuity Plan")
        print("4. Exit")
        
        choice = input("Enter your choice (1-4): ")
        
        if choice == '1':
            calculate_simple_interest()
        elif choice == '2':
            calculate_compound_interest()
        elif choice == '3':
            calculate_annuity_plan()
        elif choice == '4':
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option from the menu.")
```
